```python
import pandas as pd
import os
```

```python
def getSeconds(imlname):
    namearray = imlname.split(" ")
    
    hr = int(namearray[1].replace('h',''))
    mi = int(namearray[2].replace('m',''))
    sec = float(namearray[3].replace('s.jpg',''))
    totalSec= int(sec + mi*60 + hr*3600)
    
    return totalSec
```

```python
ListResults = []
Slno = 0
```

```python
#Here all stage 1 to be mapped with Lead
#All stage 2 to be mapped with Zinc
```

```python
#to be editted every iteration 
ImageFolder = r"C:\Users\sidda\Desktop\plots\Figure13model\Cleaned_ImageData\Run_9\Image data\Stage 2\pics"
ExcelSheet = r"C:\Users\sidda\Desktop\plots\Figure13model\runs_excel_files\R9.xlsx"

SheetName = "R9_Zinc"
Element = "Zinc"
Stage = 2
Run = 9

startHr= 22
startMn = 24
startSec = 30
start_totalSec= startSec + startMn*60 + startHr*3600
print (start_totalSec)

df = pd.read_excel(ExcelSheet, sheet_name=SheetName, usecols = 'A:B' )
#df = pd.read_csv(ExcelSheet, usecols = 'A:B' )
print(df.shape)
df.head()
```

```python
imgFileNames = [f for f in os.listdir(ImageFolder) if os.path.isfile(os.path.join(ImageFolder, f))]
for img in imgFileNames:
    Slno = Slno+1
    resultDict = {}
    
    print(img)
    timeSec = getSeconds(img)
    expTime = int(timeSec - start_totalSec)
    print(expTime)
    
    for index, row in df[df.Time == expTime ].iterrows():
        #print(row)
        recov = row['Recovery']
        print(recov)
        
        resultDict['Sl_No'] = Slno
        resultDict['FileName'] = img
        resultDict['Address'] = os.path.join(ImageFolder, img)
        resultDict['Recovery'] = recov
        resultDict['Element'] = Element
        resultDict['Stage'] = Stage
        resultDict['Run'] = Run
        resultDict['SheetName'] = SheetName
        print(resultDict)
        
        ListResults.append(resultDict)
        continue
    
    
        
    
    
```

```python
#To Get Results
dfresults = pd.DataFrame(ListResults)
dfresults.head()
```

```python
# Bringing all Run results into a csv file
os.chdir (r'C:\Users\sidda\Desktop\plots\Figure13model\mapped_data')
dfresults.to_csv('AllRunsmapped.csv', index = False)
```
